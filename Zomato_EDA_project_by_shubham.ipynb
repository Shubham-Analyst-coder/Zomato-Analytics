```python
#null value
df[df["zipcode"].isnull()]
```

```python
# drop column
df.drop('zipcode', axis=1, inplace=True)
display(df.head())
```

```python
df[df["timings"].isnull()]
```

```python
df["timings"].mode()
```

```python
#fill the null value
df["timings"]=df["timings"].fillna(df["timings"].mode()[0])
```

```python
df["cuisines"].mode()
```

```python
df["cuisines"]=df["cuisines"].fillna(df["cuisines"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df['address'].fillna('Unknown', inplace=True)
print(df['address'].isnull().sum())
```

```python
df["opentable_support"].isnull().sum()
```

```python
df["opentable_support"].median()
```

```python
df["opentable_support"]=df["opentable_support"].fillna(df["opentable_support"].median())
```

```python
#check duplicate
df.duplicated().sum()
```

```python
#drop duplicate
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
# check data clean
sns.heatmap(df.isnull())
```

```python
# Find the top 10 cities with the most restaurants
top_cities = df['city'].value_counts().nlargest(10)

# Display the result
print("Top 10 cities with the most restaurants:")
(top_cities)
```

This project involved performing an in-depth Exploratory Data Analysis (EDA) on the Indian restaurants dataset sourced from Zomato. The dataset contained over 2 lakh+ restaurant records across multiple cities, offering a comprehensive look at India‚Äôs diverse and rapidly growing food service industry.


üîç **Key Observations from the Analysis:**
Restaurant Distribution by City
The cities with the highest number of restaurants are:

Chennai, Mumbai, Bangalore, Pune, and New Delhi
These urban hubs dominate the food service market, highlighting significant customer demand and competition.

Ratings and Customer Satisfaction

A large number of restaurants have ratings between 3.5 and 4.5, which indicates generally high customer satisfaction.

The most common rating category is ‚ÄúVery Good‚Äù, followed by ‚ÄúGood‚Äù and ‚ÄúExcellent.‚Äù

Popular Cuisines

North Indian, Chinese, South Indian, and Fast Food are among the most frequently served cuisines.

Multicuisine offerings are also common, appealing to a wider customer base.

Average Cost for Two

The typical cost for two people ranges from ‚Çπ250 to ‚Çπ700, making dining out relatively affordable in most areas.

Cities like Mumbai and Bangalore show slightly higher average costs, reflecting premium locations and higher operational expenses.

Online Delivery & Takeaway Services

Restaurants that provide online delivery and takeaway options often have better aggregate ratings.

This reflects modern consumer preferences for convenience and speed, especially in urban centers.

Top Restaurant Chains

The most common chains are Domino's Pizza, Cafe Coffee Day, KFC, McDonald‚Äôs, and Subway.

These brands have built strong footprints across various cities, showing brand consistency and operational scalability.

**Data Cleaning Insights**

The dataset initially had a significant amount of missing and duplicate data.

Cleaning included removing over 1.5 lakh duplicate rows, filling in missing values using mode/median imputation, and dropping irrelevant columns like zipcode.

This step was critical to ensure data quality and analysis accuracy.
